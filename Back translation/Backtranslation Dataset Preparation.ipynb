```python
import pandas as pd
from deep_translator import GoogleTranslator
from tqdm import tqdm, tqdm_notebook

tqdm_notebook().pandas()

def back_translate_column(df, column_name, source_lang, target_lang, new_column_name):
    def back_translate(text):
        try:
            # Translate from source to target language
            translated_text = GoogleTranslator(source=source_lang, target=target_lang).translate(text)

            # Translate back to the original language
            back_translated_text = GoogleTranslator(source=target_lang, target=source_lang).translate(translated_text)

            return back_translated_text
        except Exception as e:
            print(f"Translation error for text: '{text}'. Error: {e}")
            return f"Error: Translation not found"

    df[new_column_name] = df[column_name].progress_apply(back_translate)
    return df

source_language = 'en' # Change language if needed
target_language = 'ru' # Same
new_column_name = f'Back_Translated_{source_language.upper()}_to_{target_language.upper()}'

back_translated_df = back_translate_column(df, 'Text', source_language, target_language, new_column_name)

back_translated_df.to_csv('en_russian_en.csv', encoding='utf-8', index=False)

back_translated_df.head()
```

```python
def transform_df(backtranslated_df, column_name):
    """
    Use this function to transform dataset. Do not forget to provide column name!
    """
    new_rows = pd.DataFrame({
        'Text': backtranslated_df[f'{column_name}'],
        'Sentiment': backtranslated_df['Sentiment']
    })

    df_extended = pd.concat([backtranslated_df[['Text', 'Sentiment']], new_rows], ignore_index=True)

    return df_extended
```

```python
df = transform_df(back_translated_df, 'Back_Translated_EN_to_RU')
```
